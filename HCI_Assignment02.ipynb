```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to plot any shape on a Cartesian plane
def plot_shape(shape, title="Shape"):
    x = np.append(shape[:, 0], shape[0, 0])
    y = np.append(shape[:, 1], shape[0, 1])

    fig, ax = plt.subplots(figsize=(7,7))
    ax.plot(x, y, marker='o', linewidth=2)

    # Axes at origin
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')

    # Limits & ticks
    ax.set_xlim(-7, 7)
    ax.set_ylim(-7, 7)
    ax.set_xticks(range(-6, 7))
    ax.set_yticks(range(-6, 7))

    ax.grid(True)
    ax.set_aspect('equal')
    ax.set_title(title)

    # Axis labels placed **outside numbers** to avoid overlap
    ax.text(6.2, -0.5, "+X-axis", fontsize=10)
    ax.text(-7.8, -0.5, "-X-axis", fontsize=10)
    ax.text(0.2, 6.8, "+Y-axis", fontsize=10)
    ax.text(0.2, -6.2, "-Y-axis", fontsize=10)

    plt.show()
```

```python
# Empty Cartesian plane
plot_shape(np.array([[0,0]]), "Empty Cartesian Plane")
```

```python
# Original triangle
triangle = np.array([[1,1],[3,1],[2,3]])
plot_shape(triangle, "Original Triangle")
```

```python
# Translation
translated = triangle + np.array([2,1])
plot_shape(translated, "After Translation")
```

```python
# Rotation
theta = np.radians(45)
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],[np.sin(theta),  np.cos(theta)]])
rotated = translated @ rotation_matrix.T
plot_shape(rotated, "Rotation on Translated Shape")
```

```python
# 5 Scaling
scaled = translated * np.array([1.5, 1.5])
plot_shape(scaled, "Scaling on Translated Shape")
```

```python
# Reflection
reflect_x = translated * np.array([1,-1])
plot_shape(reflect_x, "Reflection about X-axis")
reflect_y = translated * np.array([-1,1])
plot_shape(reflect_y, "Reflection about Y-axis")
```
