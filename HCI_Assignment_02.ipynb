```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plot Function
def plot_shape(shape, title="Shape", original_shape=None):
    fig, ax = plt.subplots(figsize=(7,7))

    # Original shape in background (light gray)
    if original_shape is not None:
        ox = np.append(original_shape[:, 0], original_shape[0, 0])
        oy = np.append(original_shape[:, 1], original_shape[0, 1])
        ax.plot(ox, oy, color='lightgray', linewidth=2, linestyle='--', label='Original')

    # Current shape
    x = np.append(shape[:, 0], shape[0, 0])
    y = np.append(shape[:, 1], shape[0, 1])
    ax.plot(x, y, marker='o', linewidth=2, label='Transformed')

    # Axes at origin
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')

    ax.set_xlim(-6, 6)
    ax.set_ylim(-6, 6)
    ax.set_xticks(range(-6, 7))
    ax.set_yticks(range(-6, 7))

    ax.grid(True)
    ax.set_aspect('equal')
    ax.set_title(title)

    # Axis labels
    ax.text(6, -1, "+X", fontsize=10, ha='center')
    ax.text(-6, -1, "-X", fontsize=10, ha='center')
    ax.text(0.5, 5.5, "+Y", fontsize=10)
    ax.text(0.5, -5.5, "-Y", fontsize=10)

    ax.legend()
    plt.show()
```

```python
# Empty Cartesian Plane
plot_shape(np.array([[0, 0]]), "Empty Cartesian Plane")
```

```python
# Original Shape
triangle = np.array([
    [1, 1],
    [3, 1],
    [2, 3]
])

# Original Triangle
plot_shape(triangle, "Original Triangle")
```

```python
# Translation
translated = triangle + np.array([2, 1])
plot_shape(translated, "Translation of Original Shape", original_shape=triangle)
```

```python
# Rotation
theta = np.radians(45)
rotation_matrix = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

rotated = triangle @ rotation_matrix.T
plot_shape(rotated, "Rotation of Original Shape", original_shape=triangle)
```

```python
# Scaling
scaled = triangle * np.array([1.5, 1.5])
plot_shape(scaled, "Scaling of Original Shape", original_shape=triangle)
```

```python
# Reflection about X-axis
reflect_x = triangle * np.array([1, -1])
plot_shape(reflect_x, "Reflection about X-axis", original_shape=triangle)

# Reflection about Y-axis
reflect_y = triangle * np.array([-1, 1])
plot_shape(reflect_y, "Reflection about Y-axis", original_shape=triangle)
```
